```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('../vehicles_us.csv')
df.head()

fig = px.histogram(df, x='price', nbins=50, title="Price Distribution of Vehicles")
fig.show()
```

### Intermediate conclusion (Price distribution)
Most vehicle prices are concentrated in the lower-to-mid price range, with fewer vehicles at higher prices.
The distribution is right-skewed, which suggests a small number of expensive listings act as outliers.

```python
fig3 = px.scatter(
    df,
    x='odometer',
    y='price',
    color='fuel',
    title="Odometer vs Price by Fuel Type",
    opacity=0.6
)
fig3.show()
```

### Intermediate conclusion (Odometer vs. price)
There is a general trend where higher mileage vehicles tend to have lower prices.
However, the relationship is not perfectly linear, and fuel type appears to create visible clusters or differences in price at similar mileage levels.

## Overall conclusion
In this notebook, I loaded the used vehicle advertisements dataset, explored the main variables, and visualized price patterns.
The price distribution is skewed with a majority of listings in lower-to-mid price ranges and a smaller number of high-price outliers.
Mileage generally has a negative relationship with price, and fuel type may help explain differences in pricing at similar mileage.
